```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ftp://webftp.vancouver.ca/OpenData/csv/electric_vehicle_charging_stations.csv')
df.head(5)
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
Vancouver_Coordinates = (49.2827, -123.1207)
```

```python
map = folium.Map(location = Vancouver_Coordinates, 
                 zoom_start=12,
                tiles = 'Stamen Toner')

marker_cluster = MarkerCluster().add_to(map)

for each in df[:].iterrows():
    folium.Marker(
        location = [each[1]['LATITUDE'],each[1]['LONGITUDE']],
    #popup = df.loc[each]['ADDRESS'],
    icon = folium.Icon(color='black', icon='fas fa-plug', prefix='fa'),
    ).add_to(marker_cluster),

display(map)
```
